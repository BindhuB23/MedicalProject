```python
import pandas as pd
import numpy as np
import seaborn as sn
```

```python
df=pd.read_csv("medical_cost (1).csv")
df
```

```python
df.replace("?",np.nan,inplace=True)
df.isnull().sum()
```

```python
sn.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['smoker']=df['smoker'].astype('category')
df['smoker']=df['smoker'].cat.codes
df
```

```python
df['region']=df['region'].astype('category')
df['region']=df['region'].cat.codes
df
```

```python
df['sex']=df['sex'].astype('category')
df['sex']=df['sex'].cat.codes
df.head()
```

```python
x=df.iloc[:,1:7]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

# 70:30

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import GridSearchCV
from mpl_toolkits.mplot3d import Axes3D

# Define parameter grids
param_grid_rf = {
    'n_estimators': [100, 150, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}
param_grid_lr = {
    'fit_intercept': [True, False],
    
}
param_grid_svr = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto'],
    'degree': [3, 4, 5]
}

# Initialize models
models = {
    'Random Forest': (RandomForestRegressor(), param_grid_rf),
    'Linear Regression': (LinearRegression(), param_grid_lr),
    'Support Vector Regressor': (SVR(), param_grid_svr)
}

results = {
    'Model': [],
    'Dataset': [],
    'MAE': [],
    'MSE': [],
    'RMSE': [],
    'R2': []
}

# Perform grid search and evaluate each model
for model_name, (model, param_grid) in models.items():
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='r2').fit(x_train_scaled, y_train)
    best_model = model.set_params(**grid_search.best_params_)
    best_model.fit(x_train_scaled, y_train)
    
    # Training set
    y_pred_train = best_model.predict(x_train_scaled)
    mae_train = mean_absolute_error(y_train, y_pred_train)
    mse_train = mean_squared_error(y_train, y_pred_train)
    rmse_train = np.sqrt(mse_train)
    r2_train = r2_score(y_train, y_pred_train)
    
    results['Model'].append(model_name)
    results['Dataset'].append('Training')
    results['MAE'].append(mae_train)
    results['MSE'].append(mse_train)
    results['RMSE'].append(rmse_train)
    results['R2'].append(r2_train)
    
    # Testing set
    y_pred_test = best_model.predict(x_test_scaled)
    mae_test = mean_absolute_error(y_test, y_pred_test)
    mse_test = mean_squared_error(y_test, y_pred_test)
    rmse_test = np.sqrt(mse_test)
    r2_test = r2_score(y_test, y_pred_test)
    
    results['Model'].append(model_name)
    results['Dataset'].append('Testing')
    results['MAE'].append(mae_test)
    results['MSE'].append(mse_test)
    results['RMSE'].append(rmse_test)
    results['R2'].append(r2_test)

# Print the results
for i in range(0, len(results['Model']), 2):
    print(f"Model: {results['Model'][i]}")
    print(f"  Training Set:")
    print(f"    MAE: {results['MAE'][i]}")
    print(f"    MSE: {results['MSE'][i]}")
    print(f"    RMSE: {results['RMSE'][i]}")
    print(f"    R2: {results['R2'][i]}")
    print(f"  Testing Set:")
    print(f"    MAE: {results['MAE'][i+1]}")
    print(f"    MSE: {results['MSE'][i+1]}")
    print(f"    RMSE: {results['RMSE'][i+1]}")
    print(f"    R2: {results['R2'][i+1]}")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(results['Model'][::2]))  # the label locations
width = 0.35  # the width of the bars

# Extract R2 scores for train and test sets
y_train_r2 = np.array([results['R2'][j] for j in range(0, len(results['Model']), 2)])
y_test_r2 = np.array([results['R2'][j] for j in range(1, len(results['Model']), 2)])

# Plot bars
ax.bar(x - width/2, y_train_r2, width, label='Train R2', color='blue')
ax.bar(x + width/2, y_test_r2, width, label='Test R2', color='orange')

# Annotate bars with model names
ax.set_xticks(x)
ax.set_xticklabels(results['Model'][::2], rotation=45, ha='right')

ax.set_xlabel('Model')
ax.set_ylabel('R2 Score')
ax.set_title('R2 Score Comparison of Different Models(70:30)')
ax.legend()
plt.show()
```
